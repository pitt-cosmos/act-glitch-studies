# Energy per Detector

```python
data_unf = np.genfromtxt('icecube_unf.txt')
data_f = np.genfromtxt('icecube_crf.txt')

data_unf = np.log(data_unf)
data_f = np.log(data_f)


plt.figure(figsize=(12,7))

plt.step(*data_unf, label="Unifiltered")
plt.step(*data_f,label="CR-Filtered")
plt.xlabel('$log$(Energy) $pJ$')
plt.ylabel('$log$(Events)')
plt.title('Energy per Detector 0-150 TODs (IceCube)')
plt.legend()
plt.show()
```

```python
slope_unf, intercept_unf = np.polyfit(data_unf[0][3:45],data_unf[1][3:45],1)
data_unfy = slope_unf*data_unf[0][3:45] + intercept_unf

slope_f, intercept_f = np.polyfit(data_f[0][3:20],data_f[1][3:20],1)
data_fy = slope_f*data_f[0][3:20] + intercept_f

plt.figure(figsize=(12,7))
plt.plot(data_unf[0][3:45],data_unfy,'--', label='Slope = ' + str(slope_unf))
plt.plot(data_f[0][3:20],data_fy,'--', label='Slope = ' + str(slope_f))


plt.step(*data_unf, label="Unfiltered")
plt.step(*data_f,label="CR-Filtered")
plt.xlabel('$log$(Energy) $pJ$')
plt.ylabel('$log$(Events)')
plt.title('Energy per Detector, 150 TODs (IceCube)')
plt.legend()
plt.show()
```
